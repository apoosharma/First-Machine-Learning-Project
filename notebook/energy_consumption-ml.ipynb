```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/dataquestio/project-walkthroughs/refs/heads/master/beginner_ml/teams.csv"
teams=pd.read_csv(url)
teams
```

```python
teams
```

```python
teams = teams[["team","country","year","athletes","age","prev_medals","medals"]]
```

```python
teams
```

```python
import seaborn as sns
```

```python
sns.lmplot(x='athletes',y='medals',data=teams,fit_reg=True,ci=None)
```

```python
teams.plot.hist(y="medals")
```

```python
teams[teams.isnull().any(axis=1)]
```

```python
teams = teams.dropna()
```

```python
teams
```

```python
teams.shape
```

```python
train=teams[teams["year"]<2012].copy()
```

```python
test=teams[teams["year"]>=2012].copy()
```

```python
test.shape
```

```python
train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors=["athletes","prev_medals"]
target=["medals"]
```

```python
reg.fit(train[predictors], train[target])
```

```python
predictions= reg.predict(test[predictors])
```

```python
test["predictors"]=predictions
```

```python
test
```

```python
test.loc[ test["predictors"]< 0,"predictors"] = 0
```

```python
test["predictors"]=test["predictors"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"],test["predictors"])
```

```python
error
```

```python
errors = (test["medals"] - test["predictors"]).abs()
```

```python
error_by_team = errors.groupby(test["team"]).mean()
```

```python
medals_by_team = test["medals"].groupby(test["team"]).mean()
```

```python
error_ratio =error_by_team/medals_by_team
```
